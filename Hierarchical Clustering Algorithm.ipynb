```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import zscore
```

```python
data=pd.read_csv('Cust_Spend_Data.csv')
```

```python
data.head(10)
```

```python
data_attr=data.iloc[:,2:]
```

```python
data_attr
```

```python
data_attr_scaled=data_attr.apply(zscore)
```

```python
data_attr_scaled
```

```python
sns.pairplot(data_attr_scaled,height=2,aspect=2,diag_kind='kde')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='average')
```

```python
import warnings
warnings.filterwarnings('ignore')
model.fit(data_attr_scaled)
```

```python
data_attr['labels']=model.labels_
```

```python
data_attr
```

```python
data_attr[data_attr['labels']==0]
```

```python
data_attr[data_attr['labels']==1]
```

```python
data_attr[data_attr['labels']==2]
```

```python
data=data_attr.groupby(['labels'])
```

```python
data.mean()
```

```python
#now we are going to build dendrogram
```

```python
from scipy.cluster.hierarchy import cophenet,dendrogram,linkage
```

```python
from scipy.spatial.distance import pdist #pdist is pair wise ditribution between data points
```

```python
d=linkage(data_attr_scaled,metric='euclidean',method='average')
```

```python
c, coph_dist=cophenet(d,pdist(data_attr_scaled))
```

```python
c
```

```python
plt.figure(figsize=(10,5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
dendrogram(d,leaf_rotation=15,leaf_font_size=10)
plt.tight_layout()
```
